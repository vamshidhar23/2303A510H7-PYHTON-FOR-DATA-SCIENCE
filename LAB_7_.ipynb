# **PART!!! 1**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("daily_temperature.csv")
```

```python
print(data.head())
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    data["Day"],
    data["Temperature"],
    marker='o',
    linestyle='-',
    color='b',
    label="Temperature"
)

plt.xlabel("Day of the Month")
plt.ylabel("Temperature (Â°C)")
plt.title("Daily Temperature Changes Over a Month")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.show()
```

# **Part - 2**

```python
data = pd.read_csv("sales_data.csv")
```

```python
plt.figure(figsize=(8,5))
plt.plot(
    data["Quarter"],
    data["Sales"],
    marker='o',
    linestyle='-',
    color='b',
    label="Sales"
)

peak = data.loc[data["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color='red', s=100, zorder=5, label="Peak Sales")

plt.xlabel("Quarter")
plt.ylabel("Sales ($)")
plt.title("Quarterly Sales Performance")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.show()
```

```python
print(f"Peak Sales: {peak['Quarter']} with ${peak['Sales']}")
```
